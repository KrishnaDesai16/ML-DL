#  First Step:Normality check :Data is normally distributed or not

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data['volatile acidity'].hist(figsize=(10,8),bins=10)
```

```python
#check the data is normally distributed or not 
from scipy.stats import shapiro
DataToTest=data['volatile acidity']
stat,p=shapiro(DataToTest)
print('stats=%0.2f p=%0.30f'%(stat,p))
if p>0.05:
    print('data is normally distributed')
else:
    print('Data is not normally distributed')
```

# Correlation of categorical variable - Chi square test

```python

data=pd.read_csv('https://gist.githubusercontent.com/michhar/2dfd2de0d4f8727f873422c5d959fff5/raw/ff414a1bcfcba32481e4d4e8db578e55872a2ca1/titanic.csv', sep='\t')
data.head()
```

```python
from scipy.stats import chi2_contingency
contingency_data = pd.crosstab(data['Sex'], data['Embarked'],margins = False)
stat, p,dof,expected = chi2_contingency(contingency_data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('independent categories')
else:
    print('dependent categories')
```

```python
print('mean of my_score:',np.mean(my_score))
print('mean of virat_score:',np.mean(virat_score))
```

```python
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel
```

```python
#one sample t test
ttest_1samp(my_score,1) #reject the h0 :mean of my score is 1.
```

```python
#two sample two ttest    independent samples
ttest_ind(my_score,virat_score) #reject h0 :mean of my score and virat score is different
```

```python
#paired sample ttest     son same sample with diffrent interval
meanofmy_score_tour=[23,45,56,78,3445,67,7,8,67,67]
ttest_rel(my_score,meanofmy_score_tour)
```

# Test of Stationarity - very Important for time series analysis

```python
from statsmodels.tsa.stattools import adfuller
stock_price_data = [121,131,142,121,131,142,121,131,142]
stat, p, lags, obs, crit, t = adfuller(stock_price_data)
```

```python
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Series is not Stationary')
else:
    print('Series is stationary')
```
