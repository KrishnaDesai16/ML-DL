## Target guided Ordinal Encoding
1. Ordering the lables according to target
2. Replace the labels according to the joint probability of target 0 or 1

```python
df= pd.read_csv("https://gist.githubusercontent.com/michhar/2dfd2de0d4f8727f873422c5d959fff5/raw/ff414a1bcfcba32481e4d4e8db578e55872a2ca1/titanic.csv",
                           sep='\t',usecols=['Survived','Cabin'])   
df
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df
```

```python
l=[]
data=df['Cabin'].astype(str)
for i in data:
    l.append(i[0])
    
```

```python
#df['Cabin'].astype(str).str[0]
```

```python
df['Cabin']=l
```

```python
df.head()
```

```python
pd.Categorical(df['Cabin'])
```

```python
cabin_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
cabin_labels
```

```python
cabin_labels2={k:i for i,k in enumerate(cabin_labels,0)}
```

```python
df['cabin_ordinal']=df['Cabin'].map(cabin_labels2)
```

```python
df.head(20)
```

## Probability Ratio Encoding


Probability Ratio Encoding
Probability of Survived based on Cabin--- Categorical Feature
Probability of Not Survived---1-pr(Survived)
pr(Survived)/pr(Not Survived)
Dictonary to map cabin with probability
replace with the categorical feature

```python
df= pd.read_csv("https://gist.githubusercontent.com/michhar/2dfd2de0d4f8727f873422c5d959fff5/raw/ff414a1bcfcba32481e4d4e8db578e55872a2ca1/titanic.csv",
                           sep='\t',usecols=['Survived','Cabin'])   
df
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.head()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
data=pd.DataFrame(df.groupby(['Cabin'])['Survived'].mean())
```

```python
data['Died']=1-data['Survived']
data['Prpbability_ratio_endoing']=data['Survived']/data['Died']
```

```python
s=data['Prpbability_ratio_endoing'].to_dict()
```

```python
df['Cabin_encoded']=df['Cabin'].map(s)
```
