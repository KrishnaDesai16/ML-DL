#### What if there are many categories ..select top 10 most frequant categories from each feature

```python
df=pd.read_csv('mercedes.csv',usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

```python
last10=df.X1.value_counts().head(10).index
last10=list(last10)
```

```python
last10
```

```python
for i in last10:
     df[i]=np.where(df['X1']==i,1,0)
```
